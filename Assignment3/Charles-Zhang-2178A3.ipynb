```python
# Reshape the DataFrame to have fall scores grouped by income group
fall_scores_by_income = df[['fallreadingscore', 'fallmathscore', 'fallgeneralknowledgescore', 'incomegroup']]

# Visualize scores distribution by income group
fig, ax = plt.subplots(figsize=(12, 7))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.yaxis.set_ticks_position('none')

ax.grid(color='grey', axis='y', linestyle='-', linewidth=0.25, alpha=0.5)

# Create boxplot without outliers
fall_scores_by_income.boxplot(column=['fallreadingscore', 'fallmathscore', 'fallgeneralknowledgescore'], by='incomegroup', ax=ax, showfliers=False)

ax.set_ylabel('Score')

ax.set_title('Distribution of fall scores by income group')

plt.show()
```

```python
def get_summary_statistics(dataset):

    mean = np.round(np.mean(dataset), 2)
    median = np.round(np.median(dataset), 2)
    min_value = np.round(dataset.min(), 2)
    max_value = np.round(dataset.max(), 2)
    quartile_1 = np.round(dataset.quantile(0.25), 2)
    quartile_3 = np.round(dataset.quantile(0.75), 2)

    # Interquartile range
    iqr = np.round(quartile_3 - quartile_1, 2)

    print('Min: %s' % min_value)
    print('Mean: %s' % mean)
    print('Max: %s' % max_value)
    print('25th percentile: %s' % quartile_1)
    print('Median: %s' % median)
    print('75th percentile: %s' % quartile_3)
    print('Interquartile range (IQR): %s' % iqr)
    print('Setosa summary statistics')
    print('\n\nSetosa summary statistics')
```

```python
print('\n\nspring scores summary statistics')
get_summary_statistics(spring_scores_by_income)
```

```python
print('\n\nfall scores summary statistics')
get_summary_statistics(fall_scores_by_income)
```

```python
from scipy import stats

# Population Mean
mu = 10

# Sample Size
N = len(df)  # Assuming the sample size is the same for both fall and spring scores

# Fall Scores
fall_scores = df[['fallreadingscore', 'fallmathscore', 'fallgeneralknowledgescore']].mean(axis=1)

# Spring Scores
spring_scores = df[['springreadingscore', 'springmathscore', 'springgeneralknowledgescore']].mean(axis=1)

t_stat_fall, p_val_fall = stats.ttest_1samp(a=fall_scores, popmean=mu)
print("Fall Scores - t-statistic =", t_stat_fall)
print("Fall Scores - p-value =", p_val_fall)

t_stat_spring, p_val_spring = stats.ttest_1samp(a=spring_scores, popmean=mu)
print("Spring Scores - t-statistic =", t_stat_spring)
print("Spring Scores - p-value =", p_val_spring)
```

```python
from scipy.stats import ttest_ind

# Perform two-sample t-test
t_stat, p_val = ttest_ind(fall_scores, spring_scores)

print("Two-sample t-test:")
print("t-statistic =", t_stat)
print("p-value =", p_val)
```

```python

# Assumption 1: Residuals are normally distributed (Shapiro-Wilk test)
w, pvalue = stats.shapiro(ancova_model_fall_general.resid)
print("Shapiro-Wilk test p-value:", pvalue)
```

```python
# Assumption 2: Variances are homogenous (Levene's test)
levene_test = stats.levene(df['fallgeneralknowledgescore'], df['incomegroup'])
print("Levene's test p-value:", levene_test.pvalue)
```
