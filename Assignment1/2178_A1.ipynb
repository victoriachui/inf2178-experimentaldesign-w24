```python
import pandas as pd

# Load the data
df = pd.read_excel('/content/INF2178_A1_data.xlsx')
df
```

```python
# Understanding the general idea of the dataset
df.info()
```

```python
# Removing Program ID and Program Name Columns because they are less relevant
df = df.drop(columns=['PROGRAM_ID', 'PROGRAM_NAME'])
df
```

```python
# CAPACITY_ACTUAL_BED	could be merged with CAPACITY_ACTUAL_ROOM, since its only a difference of capacity type. This is so that the NaN values cancel out
df['CAPACITY'] = df['CAPACITY_ACTUAL_BED'].fillna(df['CAPACITY_ACTUAL_ROOM'])
df
```

```python
# The same could be said for occupied beds and occupied rooms
df['OCCUPIED'] = df['OCCUPIED_BEDS'].fillna(df['OCCUPIED_ROOMS'])
df
```

```python
# Now we can drop the existing four columns for capacity and occupied
df = df.drop(columns=['CAPACITY_ACTUAL_BED', 'CAPACITY_ACTUAL_ROOM', 'OCCUPIED_BEDS', 'OCCUPIED_ROOMS'])
df
```

```python
# From this cleaned data we could also identify the occupancy rate
df['OCCUPANCY RATE'] = (df['OCCUPIED'] / df['CAPACITY'] * 100).round(2)
df
```

```python
# Check if there are still null values in the dataset
df.isnull().sum()
```

```python
# Drop these null values
df.dropna(subset=['PROGRAM_MODEL', 'OVERNIGHT_SERVICE_TYPE', 'PROGRAM_AREA'], inplace=True)
df
```

```python
# Now we have cleaned the dataset, we can now begin the data exploration
df.isnull().sum()
```

```python
# Importing libraries
import numpy as np
import seaborn as sns
import scipy.stats as stats
from scipy.stats import ttest_ind
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
# Data Exploration 1: Room and Bed Based Capacities
# First, we perform a t-test to identify whether there is a significant difference between the occupancy rates of the two capacities.
# The null hypothesis is that there is no significant difference.
room = df[df['CAPACITY_TYPE'] == 'Room Based Capacity']['OCCUPANCY RATE']
bed = df[df['CAPACITY_TYPE'] == 'Bed Based Capacity']['OCCUPANCY RATE']

t = ttest_ind(room, bed)
t
```

```python
# The null hypothesis is that there is no significant difference.
# Plotting the two types of occupancy
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.countplot(y=df['CAPACITY_TYPE'])
plt.title('Capacity Types')
plt.show()

# There's far more room-based capacity than bed-based capacity
```

```python
# Data exploration 2: Occupancy Rates
# Summary statistics for occupancy rates

def stats(df):
  mean = np.round(np.mean(df), 2)
  median = np.round(np.median(df), 2)
  min_value = np.round(df.min(), 2)
  max_value = np.round(df.max(), 2)
  quartile_1 = np.round(df.quantile(0.25), 2)
  quartile_3 = np.round(df.quantile(0.75), 2)

  print('Mean: %s' % mean)
  print('Median: %s' % median)
  print('Min Value: %s' % min_value)
  print('Max Value: %s' % max_value)
  print('25th Quartile: %s' % quartile_1)
  print('75th Quartile: %s' % quartile_3)

stats(df['OCCUPANCY RATE'])
```

```python
# A t-test to identify whether there is a significant difference between the occupancy rates depending on whether the program type is emergency or transitional
# The null hypothesis is that there is no significant difference.
transitional = df[df['PROGRAM_MODEL'] == 'Transitional']['OCCUPANCY RATE']
emergency = df[df['PROGRAM_MODEL'] == 'Emergency']['OCCUPANCY RATE']

t = ttest_ind(emergency, transitional)
t
```
